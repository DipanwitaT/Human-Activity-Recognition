```python
# importing the required module
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import rcParams
from sklearn import metrics
from sklearn.model_selection import train_test_split
%matplotlib inline
sns.set(style='whitegrid', palette='muted', font_scale=1.5)
rcParams['figure.figsize'] = 14, 8
RANDOM_SEED = 42
plt.figure(figsize=(6,4))
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams.update({'font.size': 12})
# x axis values
x = [64, 128, 192, 256, 320]
# corresponding y axis values
y = [90.14, 94.73, 99.38, 96.44, 95.32]
values = [64, 128, 192, 256, 320] 
# plotting the points
plt.plot(x, y, marker="o")
plt.ylim(88,100)
plt.xlim((64,320))
plt.xticks(x,values)
# naming the x axis
plt.xlabel('Batch Size')
# naming the y axis
plt.ylabel('Accuracy (%)')
#plt.xlim(64, 320)
# giving a title to my graph
#plt.title('My first graph!')
#plt.savefig('C:/Users/Dipanwita/OneDrive/Desktop/WISDMactivity.eps', format='eps', dpi=1200)
# function to show the plot
# zip joins x and y coordinates in pairs
for x,y in zip(x,y):

    label = "{:.2f}".format(y)

    plt.annotate(label, # this is the text
                 (x,y), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center
plt.savefig('C:/Users/Dipanwita/OneDrive/Desktop/Batchsize_vs_accuracy_uci.eps', dpi=300, bbox_inches='tight')
    
plt.show()
```
